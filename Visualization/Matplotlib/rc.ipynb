````python
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline

import numpy as np
````

````python
x = np.arange(1, 100)
y = x**2
````

````python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
ax.set_title('default settings');
````

````python
mpl.rcParams['lines.linewidth'] = 2 
mpl.rcParams['lines.linestyle'] = '--'

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y);
````

<br>
You can check all the settings you can set by this command :

````python
from cycler import cycler

mpl.rcParams['axes.prop_cycle'] = cycler(color=['red', 'green', 'blue', 'pink'])

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y);
ax.plot(x, x)
ax.plot(x, y*3)
ax.plot(x, y*5);
````

as you can see, the colors cycle through what we mentioned earlier

___

you can also modify multiple rc params at once :

````python
mpl.rc('lines', linewidth=4, linestyle='-.')

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y);
````

### Temperory rc settings

with  
mpl.rc_context

````python
with mpl.rc_context({'lines.linewidth': 2, 'lines.linestyle': ':'}) :
    fig, ax = plt.subplots(figsize=(5, 3))
    ax.plot(x, y);
````

one point, 
why the color of the plot is still red???  
it's because we set the color to red **globally** before

you can also use it as a **decorator** for your functions

````python
@mpl.rc_context({'lines.linewidth': 2, 'lines.linestyle': ':'})
def plotter_func() :
    fig, ax = plt.subplots(figsize=(5, 3))
    ax.plot(x, y);
    
plotter_func()
````

you can resotre everything to default by :

````python
mpl.rcdefaults()
````

### Using style sheets

Another way to change the visual appearance of plots is to set the rcParams in a so-called style sheet and import that style sheet with matplotlib.style.use. In this way you can switch easily between different styles by simply changing the imported style sheet. A style sheets looks the same as a matplotlibrc file, but in a style sheet you can only set rcParams that are related to the actual style of a plot. Other rcParams, like backend, will be ignored. matplotlibrc files support all rcParams. The rationale behind this is to make style sheets portable between different machines without having to worry about dependencies which might or might not be installed on another machine. For a full list of rcParams see matplotlib.rcParams. For a list of rcParams that are ignored in style sheets see matplotlib.style.use.

There are a number of pre-defined styles provided by Matplotlib. For example, there's a pre-defined style called "ggplot", which emulates the aesthetics of ggplot (a popular plotting package for R). To use this style, add:

````python
plt.style.use('ggplot')
````

````python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y);
````

### Defining your own style

you can make a file, named presentation.mplstyle, and put the following lines inside it :
```text
axes.titlesize : 24
axes.labelsize : 20
lines.linewidth : 3
lines.markersize : 10
xtick.labelsize : 16
ytick.labelsize : 16
```

````python
plt.style.use('default')
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y);
````

````python
plt.style.use('./presentation.mplstyle')

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y);
````

### Composing styles

````python
plt.style.use(['dark_background', './presentation.mplstyle'])

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y);
````

### Temporary styling

````python
with plt.style.context('dark_background') :
    fig, ax = plt.subplots(figsize=(5, 3))
    ax.plot(x, y);
````
